## Importacion de librerias para el proyecto

```python
import pandas as pd
import plotly.express as px
import csv
```

## Carga de dataset para trabajar

```python

df_a = pd.read_csv("https://raw.githubusercontent.com/ISPC-Pract-prof-2023/PracticaProfesionalizante_2023/main/data/estos%20son%20del%20link%20que%20esta%20en%20notas/tur_int_turistas_residentes_destino_serie.csv")

df_b = pd.read_csv("https://raw.githubusercontent.com/ISPC-Pract-prof-2023/PracticaProfesionalizante_2023/main/data/estos%20son%20del%20link%20que%20esta%20en%20notas/tur_int_turistas_residentes_gasto_promedio_trim_destino_serie.csv")
```

## Unificar - Visualizar Datos - Limpieza de Dataframe("df_nuevo")

```python
#unir los dataframe por id
df_nuevo = pd.merge(df_a, df_b)
```

```python
#mostrar datos de columnas y tipos de datos del nuevo dataset
df_nuevo.info()
```

```python
#revisando y contando si hay datos nulos en el dataframe
df_nuevo.isnull().sum()
```

##Visualizacion distribucion de datos

```python
#Grafico Para ver el gasto promedio por destino
fig = px.bar(df_nuevo, x = 'region_destino', y ='gasto_promedio', color = 'region_destino')
fig.show()
```
